```python
file_url = "https://raw.githubusercontent.com/maxvcbhan/DADS5001-exercise-3/refs/heads/main/listings.csv"
```

```python
import pandas as pd
import plotly.express as px
import seaborn as sns

df = pd.read_csv(file_url)
df.head()
```

## Remove outlier

```python
# Drop rows where price > 10000
df_filtered = df[df['price'] <= 3000]

import plotly.express as px

fig = px.histogram(df_filtered, x="price", color= 'price' )


fig.show()
```

```python
import plotly.express as px

fig = px.box(df_filtered, y="price")
fig.show()
```

```python
import plotly.express as px

fig = px.scatter_map(df_filtered, lat="latitude", lon="longitude",  size="price", color="price", hover_name="host_name",
                  color_continuous_scale=px.colors.cyclical.IceFire, size_max=20, zoom=10)
fig.show()
```

```python
# Group by the 'neighborhood' column and calculate the average of 'price'
avg_price_by_neighborhood = df.groupby('neighbourhood')['price'].mean().reset_index()

# Rename the columns for clarity (optional)
avg_price_by_neighborhood.columns = ['neighbourhood', 'average_price']

# Display the result
print(avg_price_by_neighborhood)
```

```python
avg_price_by_neighborhood = avg_price_by_neighborhood.sort_values(by='average_price', ascending=False)
fig = px.bar(avg_price_by_neighborhood, x="neighbourhood",
                     y="average_price",
                     )
fig.show()
```
